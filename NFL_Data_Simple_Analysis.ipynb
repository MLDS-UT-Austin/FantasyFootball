We can grab any bit of data by reading from the Github repo here: https://github.com/MLDS-UT-Austin/FantasyFootball

To get any CSV, we will use the following format:

https://raw.githubusercontent.com/MLDS-UT-Austin/FantasyFootball/main/{directory}/{filename}.csv

As an example, we can get the yearly scrimmage yard leaders of 2013 by going to the following link:

https://raw.githubusercontent.com/MLDS-UT-Austin/FantasyFootball/main/yearly_stats/2013_scrimmage.csv

```python
# Grab some plotting/graphing libraries
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
scrim2013 = pd.read_csv("https://raw.githubusercontent.com/MLDS-UT-Austin/FantasyFootball/main/yearly_stats/2013_scrimmage.csv")
```

```python
scrim2013
```

```python
px.scatter(scrim2013, x="Age", y="Total Yds_YScm", hover_name="Player", symbol="AllPro")
```

```python
px.scatter(scrim2013, x="Rushing_Yds", y="Receiving_Yds", hover_name="Player")
```
